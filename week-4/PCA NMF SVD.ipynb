```python
# Configuración de matplotlib
%matplotlib inline

# Importación de librerías

# Manipulación de datos
import numpy as np
import pandas as pd
# Graficación
import matplotlib.pyplot as plt
# Scikit learn
from sklearn.decomposition import PCA, NMF, TruncatedSVD
```

```python
# Url de la imagen 
possum_url = 'https://i.pinimg.com/originals/c0/c7/10/c0c71016e37e8c2b2de6144ce33db977.png'
```

```python
possum = plt.imread(possum_url)
print (possum.shape)
plt.imshow(possum)
```

```python
'''
Función que permite comparar la transformación de una imagen 
Utilizando los algoritmos PCA NMF & SVD
Recibe como primer parámetro la imagen a transformar y el número de componentes,
que el modelo tomará como referencia para transformar la imagen.
'''
def algorithm_comparison(image, n):
    
    # Convertir la imagen a blanco y negro
    try: 
        image = image.mean(axis=2)
    except:
        pass
    
    # Lista con los modelos a instanciar
    models = [PCA, NMF, TruncatedSVD]
    # Lista para guardar la imagen transformada
    results = []
    
    c = 0
    # Configuración del gráfico    
    plt.figure(figsize= (12,12))

    # Iteración de las instancias de los modelos
    for i in models:
        # Incremento del índice para la figura
        c+=1
        # Entrenamiento del modelo tomando la cantidad de componentes
        m = i(n)
        # Transformaciones del modeo
        d = m.fit_transform(image)
        a = m.inverse_transform(d)
        # Agregar la imagen resultante a la lista de resultados
        results.append(a)
        
        # Graficación
        plt.subplot(1,3,c)
        plt.imshow(a, cmap='gray')
        plt.axis( 'off')
        
        title = str(i).split('.')[-1].replace("'>",'')
        plt.title(title, size=15)
        
    print ('3 dimentional reduction algorithms with compressing of ')
    print ('Number of components: '+ str(n))
        
    return results
```

```python
test_1 = algorithm_comparison(possum,1)
```

```python
test_5 = algorithm_comparison(possum,5)
```

```python
test_10 = algorithm_comparison(possum,10)
```

```python
test_20 = algorithm_comparison(possum,20)
```

```python
test_50 = algorithm_comparison(possum,50)
```
